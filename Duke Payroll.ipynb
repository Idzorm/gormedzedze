```python
"""
Name: Dziedzorm Duke Kwame Benson
Email: ddkbenson@gmail.com
Project: Payroll Desktop App
Class: Python For Beginners - Nov 2020 Batch
Date: 29th December 2020
"""

import PyQt5
from PyQt5 import QtWidgets, uic
from PyQt5.QtWidgets import QMessageBox
import csv
from csv import*
import os.path
import tabulate
from tabulate import*
from datetime import datetime

#Declaring my major variables globally
employee_id = ""
employee_tin = ""
employee_ssnit = ""
employee_fname = ""
employee_lname = ""
employee_position = ""
basic_sal = ""
allowances = ""
gross_sals = ""  
tier2s = ""
PAYEs = ""
tier1s = ""
paycosts = ""

def errorMsg(a, b):
    msg = QMessageBox()
    msg.setIcon(0)
    msg.setText(a)
    msg.setInformativeText(b)
    msg.setWindowTitle("Error")
    msg.show()
    msg.exec_()
    
#Create CSV file and add Headers and save data
#If it's a new operation, a CSV will be created and headers assigned.
#If not, user entry will be assigned to their resprctive cells

def save():
    if not os.path.exists("Payroll File.csv"):
        with open("Payroll File.csv", "w", newline ='') as f:
            fieldheaders = [
                "ID", "TIN", "SSNIT", "FIRST NAME", "LAST NAME", "POSITION", "START DATE", 
                "BASIC SALARY", "ALLOWANCES", "GROSS SALARY", "NET SALARY", 
                "PAYE", "TIER 1", "TIER 2", "PAYCOST"
                           ]
            writer = csv.DictWriter(f, fieldnames = fieldheaders)
            writer.writeheader()

    if call.empID.text() != "":
        employee_id = call.empID.text()
        call.empID.setText("")

    if call.tin.text() != "":
        employee_tin = call.tin.text()
        call.tin.setText("")

    if call.ssnit.text() != "":
        employee_ssnit = call.ssnit.text()
        call.ssnit.setText("")

    if call.fname.text() != "":
        employee_fname = call.fname.text()
        call.fname.setText("")

    if call.lname.text() != "":
        employee_lname = call.lname.text()
        call.lname.setText("")

    if call.position.text() != "":
        employee_position = call.position.text()
        call.position.setText("")
        
    if call.stdate.text() != "":
        employee_stdate = call.stdate.text()
        call.stdate.setText("")

    if call.bsal.text() != "":
        basic_sal = call.bsal.text()
        call.bsal.setText("")

    if call.allowa.text() !="":
        allowances = call.allowa.text()
        call.allowa.setText("")

    if call.grosslabel.text() !="":
        gross_sals = call.grosslabel.text()
        call.grosslabel.setText("")

    if call.netlabel.text() !="":
        net_sals = call.netlabel.text()
        call.netlabel.setText("")

    if call.t2label.text() !="":
        tier2s = call.t2label.text()
        call.t2label.setText("")

    if call.payelabel.text() !="":
        PAYEs = call.payelabel.text()
        call.payelabel.setText("")

    if call.t1label.text() !="":
        tier1s = call.t1label.text()
        call.t1label.setText("")

    if call.paycostlabel.text() !="":
        paycosts = call.paycostlabel.text()
        call.paycostlabel.setText("")


    fields = [
        employee_id, employee_tin, employee_ssnit, employee_fname, employee_lname,
        employee_position, employee_stdate, basic_sal, allowances, gross_sals, 
        net_sals, tier2s, PAYEs, tier1s, paycosts
            ]

    with open("Payroll File.csv", "a", newline = "") as f:
        writer = csv.writer(f)
        writer.writerow(fields)

#Searching for entries with employee unoque ID numbers        
def search():
    find = call.empID.text()
    if find == "":
        errorMsg("", "Please Enter a Valid Employee ID")
    with open("Payroll File.csv", "r") as f:
        reader = csv.reader (f)
        header = next(reader)
        for row in reader:
            if (row[0]) == find:
                
                employee_id = row[0]
                call.empID.setText(employee_id)

                employee_tin =row[1]
                call.tin.setText(employee_tin)

                employee_ssnit = row[2]
                call.ssnit.setText(employee_ssnit)

                employee_fname = row[3]
                call.fname.setText(employee_fname)

                employee_lname = row[4]
                call.lname.setText(employee_lname)

                employee_position = row[5]
                call.position.setText(employee_position) 
                
                employee_stdate = row[6]
                call.stdate.setText(employee_stdate)

                basic_sal = row[7]
                call.bsal.setText(basic_sal)

                allowances = row[8]
                call.allowa.setText(allowances) 

                gross_sals = row[9]
                call.grosslabel.setText(gross_sals)

                net_sals = row[10]
                call.netlabel.setText(net_sals)

                tier2s = row[11]
                call.t2label.setText(tier2s)

                PAYEs = row[12]
                call.payelabel.setText(PAYEs)

                tier1s = row[13]
                call.t1label.setText(tier1s)

                paycosts = row[14]
                call.paycostlabel.setText(paycosts)

#To delete an entry. User must search first
def delete():
    lines = list()
    todelete  = call.empID.text()
    if todelete == "":
        errorMsg("No Entry!", "Kindly enter an Existing ID!")
        
#Opens Csv to read rows and then remove the row that corresponds to the ID chosen    
    with open("Payroll File.csv", "r") as rf:
        reader = csv.reader(rf)
        for row in reader:
            lines.append(row)
            for field in row:
                if field == todelete:
                    lines.remove(row)
#                 else:
#                     errorMsg("", "Can't Find ID. Check and Try Again")

#Opens another CSV and writes same data minus the row removed.
    with open("Payroll File.csv", "w", newline = "") as wf:
        writer = csv.writer(wf)
        writer.writerows(lines)                
    clear()

#To calculate the various salary components when gven the Basic and the allowances.
#if no allowances, enter zero
def compute():
    #Get basic salary from line edit widget and convert to float. then assign other variables
    basic_sal = call.bsal.text()
    allowances = call.allowa.text()
    basic_salf = float(basic_sal)
    allowancesf = float(allowances) 
    gross_sal = basic_salf + allowancesf

    #A. SSNIT Contributions variables
    tier1 = (basic_salf * 0.13)
    tier2 = (basic_salf * 0.055)

    #B. PAYE 2020
    taxable = gross_sal - tier2
    PAYE = 0
    if taxable <= 319:
        PAYE = 0
    elif taxable <= 419:
        PAYE = 0 + (0.05 * (taxable - 319))
    elif taxable <= 539:
        PAYE = 5 + (0.10 * (taxable - 419))
    elif taxable <= 3539:
        PAYE = 17 + (0.175 * (taxable - 539))
    elif taxable <= 20000:
        PAYE = 542 + (0.25 * (taxable - 3539))
    else:
        PAYE = 4657.25 + (0.3 * (taxable - 20000))

    #Total Employee Deductions, net salary and pay cost.
    deductions = tier2 + PAYE
    net_sal = gross_sal - deductions
    paycost = gross_sal + tier1    

#rounding up and stringing numbers
    #basic_sals = str(round(basic_sal,2))
    #allowas = str(round(total_allowa,2))
    gross_sals = str(round(gross_sal,2))
    net_sals = str(round(net_sal,2))
    paycosts = str(round(paycost,2))
    taxables = str(round(taxable,2))
    PAYEs = str(round(PAYE,2))
    tier1s = str(round(tier1,2))
    tier2s = str(round(tier2,2))
    deductionss = str(round(deductions,2)) 

#displaying the results onto the label widget   
    call.grosslabel.setText("GHS"+" "+gross_sals)
    call.netlabel.setText("GHS"+" "+net_sals)
    call.t2label.setText("GHS"+" "+tier2s)
    call.payelabel.setText("GHS"+" "+PAYEs) 
    call.t1label.setText("GHS"+" "+tier1s) 
    call.paycostlabel.setText("GHS"+" "+paycosts)

#Clears all data from the lineedit widgets
def clear():
    employee_id = call.empID.setText("")
    employee_tin = call.tin.setText("")
    employee_ssnit = call.ssnit.setText("")
    employee_fname = call.fname.setText("")
    employee_lname = call.lname.setText("")
    employee_position = call.position.setText("")
    employee_stdate = call.stdate.setText("")
    basic_sal = call.bsal.setText("")
    allowances = call.allowa.setText("")
    gross_sals = call.grosslabel.setText("")
    net_sals = call.netlabel.setText("")   
    tier2s = call.t2label.setText("")
    PAYEs = call.payelabel.setText("")
    tier1s = call.t1label.setText("")
    paycosts = call.paycostlabel.setText("")

#Inbult employer data plus generated employee data is output to a txt file for printing or emailing as a payslip
def printslip():
    employer_name = "Institute of IT Professionals, Gh"
    employer_tin  = "C0001234567"
    employer_erno = "20201225"
    employer_address = "No. 5 Street, Accra"
    employee_id = call.empID.text()
    employee_tin = call.tin.text()   
    employee_ssnit = call.ssnit.text()
    employee_fname = call.fname.text()
    employee_lname = call.lname.text()
    employee_position = call.position.text()
    employee_stdate = call.stdate.text()
    basic_sal = call.bsal.text()
    allowances = call.allowa.text()      
    gross_sals = call.grosslabel.text()
    net_sals = call.netlabel.text()
    tier2s = call.t2label.text()
    PAYEs = call.payelabel.text()
    tier1s = call.t1label.text()
    paycosts = call.paycostlabel.text()

    header1 = ["\n EMPLOYEE PAYSLIP"]+[" "]
    table1 = [
        ["Employer: "+ employer_name, "Address: "+employer_address], ["TIN: "+employer_tin, "ER No: "+employer_erno], 
        ["Employee ID: "+employee_id, "Employee Name: "+employee_fname+" "+employee_lname],
        ["Position: "+ employee_position, "TIN No: "+employee_tin], 
        ["Start Date: "+employee_stdate, "SSNIT No: "+employee_ssnit], 
        ["Basic Salary: GHS"+ basic_sal, "Allowances: GHS"+ allowances], 
        ["Gross Salary: "+ gross_sals, "Net Salary: "+ net_sals], ["PAYE: GHS"+ PAYEs, "Tier 1: GHS"+ tier1s], 
        ["Tier 2: "+ tier2s, "Pay Cost: "+ paycosts]
        ]
    
    print(tabulate(table1, header1, tablefmt = "grid"))
    with open(employee_fname+"-payslip.txt", "w") as f:
        f.write(tabulate(table1, header1, tablefmt = "grid"))


#Defining the app and linking the GUI file.
myapp = QtWidgets.QApplication([])
call = uic.loadUi("DukePayroll.ui")

#Assigning of Buttons and other widgets here
call.btnsave.clicked.connect(save)
call.btncompute.clicked.connect(compute)
call.btnclear.clicked.connect(clear)
call.btnprint.clicked.connect(printslip)
call.btnsearch.clicked.connect(search)
call.btndelete.clicked.connect(delete)


#Load and display the Gui app
call.show()
myapp.exec()
```
